```python
data["Ram"] = data["Ram"].str.replace("GB", " ").astype("int32")
data["Weight"] = data["Weight"].str.replace("kg", " ").astype("float32")
```

```python
def add_company (input):
    
    if input == "Samsung" or input == "Mediacom" or input == "Razer" or input == "Microsoft" or input == "Vero" or input == "Xiaomi" or input == "Chuwi" or input == "Fujitsu" or input == "Google" or input == "LG" or input == "Huawei":
        return "Other"
    else:
        return input

data["Company"] = data["Company"].apply(add_company)
    
```

```python
data["Touch Screen"] = data["ScreenResolution"].apply(lambda x:1 if "Touchscreen" in  x else 0)
data["Ips"] = data["ScreenResolution"].apply(lambda x:1 if "IPS" in x else 0)
```

```python
data.head(2)
```

```python
data["cpu_name"] = data["Cpu"].apply(lambda x: " ".join(x.split()[0:3]))
```

```python
def set_processor(name):
    if name == "Intel Core i7" or name == "Intel Core i5" or name == "Intel Core i3":
        return name
    else:
        if name.split()[0] == "AMD":
            return "AMD"
        else:
            return "Other"

data["cpu_name"] = data["cpu_name"].apply(set_processor)
```

```python
data["cpu_name"].value_counts()
```

```python
len(data["Gpu"].value_counts())
```

```python
data["gpu_name"] = data["Gpu"].apply(lambda x : " " . join(x.split()[0:1]))
```

```python
data["gpu_name"].value_counts()
```

```python
data = data[data['gpu_name'] != 'ARM']
```

```python
data.shape
```

```python
data["OpSys"].value_counts()
```

```python
def set_os(inpt):
    if inpt == "Windows 10" or inpt == "Windows 7" or inpt == "Windows 10 S":
        return "Windows"
    elif inpt == "macOS" or inpt == "Mac OS X":
        return "Mac"
    elif inpt == "Linux":
        return inpt
    else :
        return "Other"

data["OpSys"] = data["OpSys"].apply(set_os)
```

```python
data["OpSys"].value_counts()
```

```python
data.head(3)
```

```python
data = data.drop(columns = ['laptop_ID', 'Inches', 'Product', 'ScreenResolution', 'Cpu', 'Gpu'])
```

```python
data.head()
```

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
data = data.astype(int)
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
X = data.drop('Price_euros', axis = 1)
y = data['Price_euros']
```

```python
X.shape, y.shape
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.25)
```

```python
X_train.shape, X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test,  y_test)
    print(str(model) + '----> '+ str(acc))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

lr = LinearRegression()
model_acc(lr)

lasso = Lasso()
model_acc(lasso)

dtr = DecisionTreeRegressor()
model_acc(dtr)

rfr = RandomForestRegressor()
model_acc(rfr)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {

    "n_estimators" : [10, 50, 100],
    "criterion" : ['squared_error', 'absolute_error', 'poisson']
}

grid_obj = GridSearchCV(estimator = rfr, param_grid = parameters)

grid_fit = grid_obj.fit(X_train, y_train)

best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test, y_test)
```
